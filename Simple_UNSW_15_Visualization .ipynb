```python

#  This piece of code shows you only visualization the type of attacs and number of each one 
#  I did not use all the Features only the target (attack_cat)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv ('UNSW_NB15_training-set.csv')
# Remove the unnessecary columns 
data_clean = data.drop(columns=['id','ct_src_ltm'])
# Check for missng valuse 
data_clean = data_clean.dropna(subset=['attack_cat'])

sns.set(style="whitegrid")

# Count of Each Attack Category 
attack_counts = data_clean['attack_cat'].value_counts()

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=attack_counts.index, y=attack_counts.values, palette='viridis')

plt.title(' Number of Each Attack Type', fontsize=16)
plt.xlabel('Attack Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

data_clean.head(5)
```
